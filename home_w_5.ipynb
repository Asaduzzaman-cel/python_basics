```python
def format_current_weather(api_data):
    city = api_data["requested_city"]
    temp = api_data["current_data"]["temperature"]
    unit = api_data["current_data"]["unit"]
    condition = api_data["current_data"]["weather_condition"]

    return f"In {city}, the current temperature is {temp} {unit} and the conditions are {condition}."
api_data_weather = {
    "status": "success",
    "requested_city": "Jakarta, ID",
    "current_data": {
        "temperature": 30.2,
        "unit": "Celsius",
        "weather_condition": "Hazy Sunshine",
        "humidity_percent": 82,
        "wind_speed_kph": 8
    }
}

print(format_current_weather(api_data_weather))
```

```python
def check_rain_alert(api_data, day_index):
    forecast = api_data["forecast_data"][day_index]

    day = forecast["day"]
    rain_chance = forecast["precipitation_chance_percent"]

    if rain_chance >= 50:
        return f"ALERT: High chance of rain {day} ({rain_chance})%! Bring an umbrella."
    else:
        return f"Forecast looks clear for {day} ({rain_chance}%)."
api_data_forecast = {
    "status": "success",
    "requested_city": "Singapore, SG",
    "forecast_data": [
        {
            "day": "tomorrow",
            "high_temp": 32,
            "low_temp": 25,
            "precipitation_chance_percent": 80
        },
        {
            "day": "the day after tomorrow",
            "high_temp": 30,
            "low_temp": 24,
            "precipitation_chance_percent": 20
        }
    ]
}

print(check_rain_alert(api_data_forecast, 0))
print(check_rain_alert(api_data_forecast, 1))
```
