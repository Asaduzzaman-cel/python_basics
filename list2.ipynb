```python
#There are 2 rules to change item from a list.
# no.1= you can change by specific refer index no.
list=["apple","banana","cherry","mango","jackfruit","orange"]
print(list)
```

```python
list[1]="pineapple"
print(list)
```

```python
list=["apple","banana","cherry","mango","jackfruit","orange"]
list[1:5]="watermelon","kiwi","grapes"
print(list)
```

```python
list=["apple","banana","cherry","mango","jackfruit","orange"]
list[1:3]="watermelon","kiwi","grapes","pepia" # যখন তুমি 1:3 করবে পাইথন মনে করবে ইনডেক্স নং 1থেকে 3 এর মধ্যে পরিবর্তন হবে। এই তিনটার মধ্যে যদি
print(list) # ৫টি এলিমেন্ট দেওয়া হয় ৫টিই নিবেএবং লিষ্ট বড় করে দিবে। যদি ৩টির মধ্যে ২টি দাও তাহলে ২ নিবে এবং লিষ্টর এলিমেন্ট সংখ্যা কমে যাবে।
```

```python
# Add item in list
# There are two rules for add item in list .append() .insert()
list=["apple","banana","cherry"]
list.append("mango")# In this way item will append at last of list
print(list)

```

```python
list.insert(0,"orenge")
print(list)
```

```python
# লিষ্টের মধ্ যেমন আইটেম যোগ করা যায় তেমনি দুটি লিসট কেও যোগ করা যায়। by extend
list1=["apple","banana","cherry"]
list2=["mango","jackfruit","orange"]
list1.extend(list2)
print(list1)
```

```python
# লিষ্টের মধ্ যেমন আইটেম যোগ করা যায় তেমনি দুটি লিসট কেও যোগ করা যায়। + ব্যবহার করে
list1=["apple","banana","cherry"]
list2=["mango","jackfruit","orange"]
list3=list1+list2
print(list3)
```
