```python
# Data Cleaner Agent
#●Concept: String Methods (lower, strip, split).
#●Task: Given a "messy" string like " DaTa_ScIeNcE ", write a program to:
#1.Remove leading/trailing spaces.
#2.Convert it to lowercase.
#3.Replace underscores with spaces.
#Print the clean version: "data science"
data=" DaTa_ScIeNcE " #data নামে ভেরিয়েবলে একটি মেসি স্ট্রিং রাখা হলো। এতে শুরু ও শেষ—দুই জায়গাতেই স্পেস আছে।
data=data.strip() #strip() শুরুর এবং শেষের বাড়তি স্পেস সরিয়ে দেয়। ফলে " DaTa_ScIeNcE " → "DaTa_ScIeNcE"
data=data.lower() #পুরো স্ট্রিং ছোট হাতের অক্ষরে রূপান্তর হয়। "DaTa_ScIeNcE" → "data_science"
data=data.replace("_"," ") #আন্ডারস্কোর _ কে স্পেস " " দিয়ে বদলে দেয়। "data_science" → "data science"
print(data) #বাড়তি স্পেস,বড় হাত–ছোট হাত অক্ষর এলোমেলো,আন্ডারস্কোর / ড্যাশ / ভুল চিহ্ন,ভুল ফরম্যাট,অপ্রয়োজনীয় ক্যারেক্টার,বানান বা কেস মিক্সড,যেটা “clean”, “standard” ডেটা না
```

```python
from ast import Return
# Rule-Based Number Classifier
# ●Concept: Conditional Logic (if/elif/else), Modulo Operator.
# ●Task: Ask the user for an integer. Write a logic flow that prints:
# ○"Even and Prime" (if applicable - Hint: 2 is the only even prime)
# ○"Even and Composite"
# ○"Odd and Prime"
# ○"Odd and Composite"
# Asking for integer input
n = int(input("Enter an integer: "))

# Function to check prime numbers
def is_prime(num):
    if num < 2:
        return False
    for i in range(2, int(num**0.5) + 1):
        if num % i == 0:
            return False
    return True

# Check even or odd
if n % 2 == 0:
    # It's even
    if is_prime(n):
        print("Even and Prime")
    else:
        print("Even and Composite")
else:
    # It's odd
    if is_prime(n):
        print("Odd and Prime")
    else:
        print("Odd and Composite")


```
